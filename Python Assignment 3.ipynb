```python
secret_number = random.randint(1, 10)
attempt = 3
print("welcome to number guessing game")
print("guess a number between 1 and 10 and you have three attempts")
while attempt > 0:
    guess = int(input("enter your numbers: "))
    if guess<1 or guess>10:
        print("invalid number . please enter a number between 1 and 10")
        continue
    if guess == secret_number:
        print("you have guessed the correct number")
        break
    elif guess > secret_number:
        print("plz guess a small number")
    else:
        print("plz guess a large number")
        
```

```python
def calculate_bmi(weight,height):
    BMI = weight  / (height **2)
    return BMI
weight = float(input("Enter your weight in kg: "))
height = float(input("Enter your height in meters: "))
BMI = calculate_bmi(weight,height)
print("Your BMI is:",BMI)            
```
