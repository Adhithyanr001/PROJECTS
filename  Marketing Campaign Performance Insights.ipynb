```python
plt.scatter(df['Acquisition_Cost'],df['ROI'])
plt.xlabel('Acquisition_Cost')
plt.ylabel('ROI')
plt.title('Relationship between Acquisition_Cost and ROI')
plt.show()
 
```

```python
sns.barplot(data=df,x= 'Conversion_Rate',y= 'Channel_Used',hue = 'Campaign_Type',estimator='mean')
plt.show()
```
